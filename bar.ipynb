# Bar plot

## Grouped bar plot

```python
import numpy as numpy

from matplotlib import pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator

np.random.seed(19260817)

x = ['A111', 'A112', 'A113', 'A114', 'A115', 'A116', 'A117', 'A118', 'A119', 'A120', 'A121', 'A122', 'A123']
y_a = np.random.rand(13) * 120
y_b = np.random.rand(13) * 120

location = np.arange(len(x))
width = 0.4

fig, ax = plt.subplots(figsize=(9, 3))
rect_hive = ax.bar(location - 0.5 * width, y_a, width, label='Label A', hatch='\\\\', color='none', edgecolor='black')
rect_ghive = ax.bar(location + 0.5 * width, y_b, width, label='Label B', hatch='//', color='none', edgecolor='black')

ax.set_xlabel('X label', fontsize=20)
ax.set_ylabel('Y label', fontsize=20)
y_major_locator = MultipleLocator(40)
ax.yaxis.set_major_locator(y_major_locator)
ax.set_xticks(np.arange(0, 13), x, fontsize=14)
ax.set_yticks([0, 30, 60, 90, 120], ['0', '30', '60', '90', '120'], fontsize=14)
ax.tick_params(direction='in', axis='both')
ax.legend(fontsize=14)
plt.savefig('./fig/grouped_bar.pdf', bbox_inches='tight', transparent=True)
```
