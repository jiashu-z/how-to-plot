# Plotting in Python

在学术论文中，我们常常需要绘制大量的点图，线图和柱图来呈现实验结果，好的作图方法允许我们在更好地呈现实验结果的同时提高工作的效率。我们有许多方式绘制论文中所需要的点图，线图和柱图，如通过Office软件，MATLAB和Origin。在本次学术论坛中，我们介绍一种跨平台的作图方法，使用Python进行作图。我们结合一些例子，向大家展示如何使用Python进行作图以及如何微调图。

The two examples in this file are from Zhang, Yingqiang, et al. "Towards cost-effective and elastic cloud database deployment via memory disaggregation." Proceedings of the VLDB Endowment 14.10 (2021): 1900-1912.

## 点图

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_moons

np.random.seed(19260817)

data = make_moons(100)

x, y = zip(*data[0])
x = np.array(x)
y = np.array(y)
labels = data[1]

# Create a figure.
plt.figure()

# Plot the points.
plt.scatter(x, y)

# In this case, we use the default color and style and everything.
```

```python
# What if we want to change the size of the marker?
# We can pass a single value, or an array of values which specifies the size of each point.
# For example, here we make half with size 10 and half with size 50.
sizes = np.array([10] * 50 + [100] * 50)

plt.figure()
plt.scatter(x, y, s=sizes)
```

```python
# Then, we want to change the color of the points.
# Let's make the color black.
plt.figure()
plt.scatter(x, y, s=sizes, c='k')
```

```python
# What if we only want the color on the edges?
plt.figure()
plt.scatter(x, y, s=sizes, c='none', edgecolors='k')
```

```python
# Then, lets change the marker.
# Let's use square.
plt.figure()
plt.scatter(x, y, s=sizes, c='none', edgecolors='k', marker='s')
```

```python
# Let's make it more complex.
# We group the points and legends.
plt.figure()
plt.scatter(x[np.where(labels == 0)], y[np.where(labels == 0)], s=100, c='none', edgecolors='b', marker='s',
            label='Label A')
plt.scatter(x[np.where(labels == 1)], y[np.where(labels == 1)], s=100, c='none', edgecolors='k', marker='o',
            label='Label B')
plt.legend()

# Apparently, the legends are very small.
```

```python
# Let's make the legends larger.
plt.figure()
plt.scatter(x[np.where(labels == 0)], y[np.where(labels == 0)], s=100, c='none', edgecolors='b', marker='s',
            label='Label A')
plt.scatter(x[np.where(labels == 1)], y[np.where(labels == 1)], s=100, c='none', edgecolors='k', marker='o',
            label='Label B')
plt.legend(fontsize=16)

# But the legends do not look very beautiful.
# We want two columns with narrower space, better position and red texts.
```

```python
# Let's adjust the legends.

plt.figure()
plt.scatter(x[np.where(labels == 0)], y[np.where(labels == 0)], s=100, c='none', edgecolors='b', marker='s',
            label='Label A')
plt.scatter(x[np.where(labels == 1)], y[np.where(labels == 1)], s=100, c='none', edgecolors='k', marker='o',
            label='Label B')
plt.legend(fontsize=16, markerscale=0.8, markerfirst=False, frameon=False, ncol=2, loc=[0.15, 1])
```

```python
# Something more?

plt.figure()
plt.scatter(x[np.where(labels == 0)], y[np.where(labels == 0)], s=100, c='none', edgecolors='b', marker='s',
            label='Label A')
plt.scatter(x[np.where(labels == 1)], y[np.where(labels == 1)], s=100, c='none', edgecolors='k', marker='o',
            label='Label B')
plt.legend(fontsize=16, labelcolor='grey', markerscale=0.8, markerfirst=False, frameon=False, ncol=2, loc=[0.25, 1],
           handlelength=0.5, columnspacing=1)
```

```python
# Or maybe you want only one column in the legends.

plt.figure()
plt.scatter(x[np.where(labels == 0)], y[np.where(labels == 0)], s=100, c='none', edgecolors='b', marker='s',
            label='Label A')
plt.scatter(x[np.where(labels == 1)], y[np.where(labels == 1)], s=100, c='none', edgecolors='k', marker='o',
            label='Label B')
plt.legend(fontsize=16, markerscale=0.8, markerfirst=False, frameon=False, handlelength=0.5, labelspacing=0.4)
```

```python
# We also notice that the x and y limits are not ideal.

plt.figure()
plt.scatter(x[np.where(labels == 0)], y[np.where(labels == 0)], s=100, c='none', edgecolors='b', marker='s',
            label='Label A')
plt.scatter(x[np.where(labels == 1)], y[np.where(labels == 1)], s=100, c='none', edgecolors='k', marker='o',
            label='Label B')
plt.legend(fontsize=16, markerscale=0.8, markerfirst=False, frameon=False, handlelength=0.5, labelspacing=0.4)
plt.xlim(-1, 2)
plt.ylim(-0.5, 1)
```

```python
# Notice the markers on edges.

plt.figure()
plt.scatter(x[np.where(labels == 0)], y[np.where(labels == 0)], s=100, c='none', edgecolors='b', marker='s',
            label='Label A', clip_on=False)
plt.scatter(x[np.where(labels == 1)], y[np.where(labels == 1)], s=100, c='none', edgecolors='k', marker='o',
            label='Label B', clip_on=False)
plt.legend(fontsize=16, markerscale=0.8, markerfirst=False, frameon=False, handlelength=0.5, labelspacing=0.4)
plt.xlim(-1, 2)
plt.ylim(-0.5, 1)
```

```python
# Now we have to add the x and y labels.

plt.figure()
plt.scatter(x[np.where(labels == 0)], y[np.where(labels == 0)], s=100, c='none', edgecolors='b', marker='s',
            label='Label A', clip_on=False)
plt.scatter(x[np.where(labels == 1)], y[np.where(labels == 1)], s=100, c='none', edgecolors='k', marker='o',
            label='Label B', clip_on=False)
plt.legend(fontsize=16, markerscale=0.8, markerfirst=False, frameon=False, handlelength=0.5, labelspacing=0.4)
plt.xlim(-1, 2)
plt.ylim(-0.5, 1)
plt.xlabel('X label')
plt.ylabel('Y label')
```

```python
# The label font size is too small!

plt.figure()
plt.scatter(x[np.where(labels == 0)], y[np.where(labels == 0)], s=100, c='none', edgecolors='b', marker='s',
            label='Label A', clip_on=False)
plt.scatter(x[np.where(labels == 1)], y[np.where(labels == 1)], s=100, c='none', edgecolors='k', marker='o',
            label='Label B', clip_on=False)
plt.legend(fontsize=16, markerscale=0.8, markerfirst=False, frameon=False, handlelength=0.5, labelspacing=0.4)
plt.xlim(-1, 2)
plt.ylim(-0.5, 1)
plt.xlabel('X label', fontsize=14)
plt.ylabel('Y label', fontsize=14, labelpad=20)
```

```python
# Now let's do something to the rulers and ticks.

plt.figure()
plt.scatter(x[np.where(labels == 0)], y[np.where(labels == 0)], s=100, c='none', edgecolors='b', marker='s',
            label='Label A', clip_on=False)
plt.scatter(x[np.where(labels == 1)], y[np.where(labels == 1)], s=100, c='none', edgecolors='k', marker='o',
            label='Label B', clip_on=False)
plt.legend(fontsize=14, markerscale=0.8, markerfirst=False, frameon=False, handlelength=0.5, labelspacing=0.4)
plt.xlim(-1, 2)
plt.ylim(-0.5, 1)
plt.xlabel('X label', fontsize=14)
plt.ylabel('Y label', fontsize=14)
plt.ylabel('Y label', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(np.arange(-0.5, 1.1, 0.3), ['a ', 'b ', 'c ', 'd ', 'e ', 'f '], fontsize=14)
plt.tick_params(axis='both', direction='in')
```

```python
# Let's make a comparison.

plt.figure()
plt.scatter(x, y)

plt.figure()
plt.scatter(x[np.where(labels == 0)], y[np.where(labels == 0)], s=100, c='none', edgecolors='b', marker='s',
            label='Label A', clip_on=False)
plt.scatter(x[np.where(labels == 1)], y[np.where(labels == 1)], s=100, c='none', edgecolors='k', marker='o',
            label='Label B', clip_on=False)
plt.legend(fontsize=14, markerscale=0.8, markerfirst=False, frameon=False, handlelength=0.5, labelspacing=0.4)
plt.xlim(-1, 2)
plt.ylim(-0.5, 1)
plt.xlabel('X label', fontsize=14)
plt.ylabel('Y label', fontsize=14)
plt.ylabel('Y label', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(np.arange(-0.5, 1.1, 0.3), ['a ', 'b ', 'c ', 'd ', 'e ', 'f '], fontsize=14)
plt.tick_params(axis='both', direction='in')
```

```python
# Now we want to change the size of the figure.

plt.figure(figsize=[5, 5])
plt.scatter(x[np.where(labels == 0)], y[np.where(labels == 0)], s=100, c='none', edgecolors='b', marker='s',
            label='Label A', clip_on=False)
plt.scatter(x[np.where(labels == 1)], y[np.where(labels == 1)], s=100, c='none', edgecolors='k', marker='o',
            label='Label B', clip_on=False)
plt.legend(fontsize=14, markerscale=0.8, markerfirst=False, frameon=False, handlelength=0.5, labelspacing=0.4)
plt.xlim(-1, 2)
plt.ylim(-0.5, 1)
plt.xlabel('X label', fontsize=14)
plt.ylabel('Y label', fontsize=14)
plt.ylabel('Y label', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(np.arange(-0.5, 1.1, 0.3), ['a ', 'b ', 'c ', 'd ', 'e ', 'f '], fontsize=14)
plt.tick_params(axis='both', direction='in')
```

```python
# Now we want to change the size of the figure.

plt.figure(figsize=[6.4, 4.8], tight_layout={'pad': 0})
plt.scatter(x[np.where(labels == 0)], y[np.where(labels == 0)], s=100, c='none', edgecolors='b', marker='s',
            label='Label A', clip_on=False)
plt.scatter(x[np.where(labels == 1)], y[np.where(labels == 1)], s=100, c='none', edgecolors='k', marker='o',
            label='Label B', clip_on=False)
plt.legend(fontsize=14, markerscale=0.8, markerfirst=False, frameon=False, handlelength=0.5, labelspacing=0.4)
plt.xlim(-1, 2)
plt.ylim(-0.5, 1)
plt.xlabel('X label', fontsize=14)
plt.ylabel('Y label', fontsize=14)
plt.ylabel('Y label', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(np.arange(-0.5, 1.1, 0.3), ['a ', 'b ', 'c ', 'd ', 'e ', 'f '], fontsize=14)
plt.tick_params(axis='both', direction='in')
# plt.savefig('fig/scatter.pdf', transparent=True)
```

## 线图

我们尝试画出来下图的效果

![](fig/line-example.png)

```python
# First, we make the dataset and plot the baseline.

x1 = np.linspace(0, 10, 40)
y1 = np.sin(x1)

x2 = np.linspace(0, 10, 30)
y2 = np.cos(x2)

x3 = np.linspace(0, 10, 20)
y3 = np.sin(x3 + 5) * 2 - 1

x4 = np.linspace(0, 10, 10)
y4 = -np.sin(x4 + 5) * 2 - 1

plt.figure()
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x3, y3)
plt.plot(x4, y4)
```

```python
# Pick some colors! https://matplotlib.org/3.5.0/gallery/color/named_colors.html

fig = plt.figure(figsize=[6.4, 4.8], dpi=100)
ax = fig.add_subplot(111)
ax.plot(x1, y1, 's-', label='Line 1', color='red', markeredgecolor='red', markerfacecolor='none', markeredgewidth=2,
        linewidth=2.5, clip_on=False, markersize=8)
ax.plot(x2, y2, 'd--', label='Line 2', color='deepskyblue', markeredgecolor='deepskyblue', markerfacecolor='none',
        linewidth=2.5, clip_on=False, markersize=8)
ax.plot(x3, y3, '^-.', label='Line 3', color='orange', markeredgecolor='orange', markerfacecolor='none', linewidth=2.5,
        clip_on=False, markersize=8)
ax.plot(x4, y4, ':', label='Line 4', color='green', markeredgecolor='green', markerfacecolor='none', linewidth=2.5,
        clip_on=False, markersize=8)
ax.legend(ncol=4, prop={'weight': 'heavy', 'size': 14}, frameon=False, handlelength=1, columnspacing=1, loc=[0.1, 1])
ax.set_xlim(0, 10)
ax.set_ylim(-3, 1)
ax.set_xlabel('X label $\mathcal{X}$', fontsize=14, weight='heavy')
ax.set_ylabel('Y label $\mathcal{Y}$', fontsize=14, style='italic', weight='heavy')
ax.set_xticks(np.arange(0, 11, 1), ['0', '', '2', '', '4', '', '6', '', '8', '', '10'], fontsize=14, weight='heavy')
ax.set_yticks([-3, -2, -1, 0, 1], ['-3', '-2', '-1', '0', '1'], fontsize=14, weight='heavy')
ax.tick_params(axis='both', direction='in', width=2)
ax.tick_params(axis='x', size=5)
ax.grid(axis='y', linestyle='--')
for axis in ax.spines:
    ax.spines[axis].set_linewidth(2)
fig.tight_layout(pad=0, w_pad=0, h_pad=0)
# plt.savefig('fig/line.pdf', transparent=True)
```

## 柱图

```python
# First, we make the dataset and plot the baseline.

x1 = np.arange(0, 13, 1)
y1 = np.array([15.33, 15.48, 12.33, 97.54, 87.62, 79.71, 97.25, 90.75, 83.68, 22.05, 124.54, 109.84, 109.04])

x2 = np.arange(0, 13, 1)
y2 = np.array([14.3, 14.72, 11.83, 43.57, 45.59, 41.83, 59.9, 40.76, 41.17, 16.17, 60.47, 59.98, 63.05])

x3 = np.arange(0, 13, 1)
y3 = np.random.randint(0, 120, 13)

x_labels = ['Q1.1', 'Q1.2', 'Q1.3', 'Q2.1', 'Q2.2', 'Q2.3', 'Q3.1', 'Q3.2', 'Q3.3', 'Q3.4', 'Q4.1', 'Q4.2', 'Q4.3']

fig = plt.figure()
ax = fig.add_subplot(111)

width = 0.3
ax.bar(x1 - width, y1, width)
ax.bar(x2, y2, width)
ax.bar(x3 + width, y3, width)
```

```python
# Let's make this figure look more professional.

fig = plt.figure(figsize=[9, 3])
ax = fig.add_subplot(111)

width = 0.3
ax.bar(x1 - width, y1, width, label='Label A', hatch='...', color='aliceblue', edgecolor='black')
ax.bar(x2, y2, width, label='Label B', hatch='///', color='lightskyblue', edgecolor='black')
ax.bar(x3 + width, y3, width, label='Label C', hatch='\\\\\\', color='royalblue', edgecolor='black')
ax.legend(fontsize=14, ncol=3, markerscale=0.8, markerfirst=False, frameon=False, handlelength=0.5, columnspacing=1,
          loc=[0.3, 1])

ax.set_xlabel('X label', fontsize=14)
ax.set_ylabel('Y label $10^{-1}$ sec', fontsize=14)
ax.set_xticks(x1, x_labels, fontsize=14)
ax.set_yticks([0, 30, 60, 90, 120], ['0', '30', '60', '90', '120'], fontsize=14)
ax.set_xlim(-1, 13)
ax.set_ylim(0, 130)
ax.tick_params(axis='both', direction='in')
ax.tick_params(axis='x', size=0)

ax2 = ax.twinx()
ax2.set_ylim(0, 130)
ax2.set_yticks([0, 30, 60, 90, 120], [''] * 5)
ax2.tick_params(axis='y', direction='in')

fig.tight_layout(pad=0)
# plt.savefig('fig/bar.pdf', transparent=True)
```

## 多图

我们尝试画出来下图的效果

![](fig/multi-example.png)

为什么要在这里画出来多个图，而不是分别画出来多个图然后拼在一起？
因为不同的图的坐标轴大小和取值范围可能不一样，在LaTex里面手动调整图的大小的话对齐比较麻烦


```python
x_labels = ['Q1.1', 'Q1.2', 'Q1.3', 'Q2.1', 'Q2.2', 'Q2.3', 'Q3.1', 'Q3.2']

x1 = np.arange(0, 8, 1)
y1 = np.random.randint(0, 120, 8)

x2 = np.arange(0, 8, 1)
y2 = np.random.randint(0, 120, 8)

x3 = np.arange(0, 8, 1)
y3 = np.random.randint(0, 120, 8)

x4 = np.arange(0, 8, 1)
y4 = np.random.randint(0, 256, 8)

x5 = np.arange(0, 8, 1)
y5 = np.random.randint(0, 256, 8)

x6 = np.arange(0, 8, 1)
y6 = np.random.randint(0, 256, 8)
```

```python
fig = plt.figure(figsize=[10, 6])

width = 0.3

ax = fig.add_subplot(221)

ax.bar(x1 - width, y1, width, label='Label A', hatch='...', color='aliceblue', edgecolor='black')
ax.bar(x2, y2, width, label='Label B', hatch='///', color='lightskyblue', edgecolor='black')
ax.bar(x3 + width, y3, width, label='Label C', hatch='\\\\\\', color='royalblue', edgecolor='black')
# ax.legend(fontsize=14, ncol=3, markerscale=0.8, markerfirst=False, frameon=False, handlelength=0.5, columnspacing=1,
#           loc=[0.3, 1])
ax.set_xlabel('X label', fontsize=14)
ax.set_ylabel('Y label $10^{-1}$ sec', fontsize=14)
ax.set_xticks(x1, x_labels, fontsize=14)
ax.set_yticks([0, 30, 60, 90, 120], ['0', '30', '60', '90', '120'], fontsize=14)
ax.set_xlim(-1, 8)
ax.set_ylim(0, 120)
ax.tick_params(axis='both', direction='in')
ax.tick_params(axis='x', size=0)
ax.set_title('(a) Subplot 1', y=-0.4, fontsize=14)

ax2 = ax.twinx()
ax2.set_ylim(0, 120)
ax2.set_yticks([0, 30, 60, 90, 120], [''] * 5)
ax2.tick_params(axis='y', direction='in')

ax = fig.add_subplot(222)

ax.bar(x4 - width, y4, width, label='Label A', hatch='...', color='aliceblue', edgecolor='black')
ax.bar(x5, y5, width, label='Label B', hatch='///', color='lightskyblue', edgecolor='black')
ax.bar(x6 + width, y6, width, label='Label C', hatch='\\\\\\', color='royalblue', edgecolor='black')
# ax.legend(fontsize=14, ncol=3, markerscale=0.8, markerfirst=False, frameon=False, handlelength=0.5, columnspacing=1,
#           loc=[0.3, 1])
ax.set_xlabel('X label', fontsize=14)
ax.set_ylabel('Y label $10^{-1}$ sec', fontsize=14)
ax.set_xticks(x1, x_labels, fontsize=14)
ax.set_yticks([0, 50, 100, 150, 200, 250], ['0', '50', '100', '150', '200', '250'], fontsize=14)
ax.set_xlim(-1, 8)
ax.set_ylim(0, 256)
ax.tick_params(axis='both', direction='in')
ax.tick_params(axis='x', size=0)

ax2 = ax.twinx()
ax2.set_ylim(0, 256)
ax2.set_yticks([0, 50, 100, 150, 200, 250], [''] * 6, fontsize=14)
ax2.tick_params(axis='y', direction='in')
ax.set_title('(b) Subplot 2', y=-0.4, fontsize=14)

ax = fig.add_subplot(212)

ax.bar(x4 - width, y4, width, label='Label A', hatch='...', color='aliceblue', edgecolor='black')
ax.bar(x5, y5, width, label='Label B', hatch='///', color='lightskyblue', edgecolor='black')
ax.bar(x6 + width, y6, width, label='Label C', hatch='\\\\\\', color='royalblue', edgecolor='black')
# ax.legend(fontsize=14, ncol=3, markerscale=0.8, markerfirst=False, frameon=False, handlelength=0.5, columnspacing=1,
#           loc=[0.3, 1])
ax.set_xlabel('X label', fontsize=14)
ax.set_ylabel('Y label $10^{-1}$ sec', fontsize=14)
ax.set_xticks(x1, x_labels, fontsize=14)
ax.set_yticks([0, 50, 100, 150, 200, 250], ['0', '50', '100', '150', '200', '250'], fontsize=14)
ax.set_xlim(-1, 8)
ax.set_ylim(0, 256)
ax.tick_params(axis='both', direction='in')
ax.tick_params(axis='x', size=0)

ax2 = ax.twinx()
ax2.set_ylim(0, 256)
ax2.set_yticks([0, 50, 100, 150, 200, 250], [''] * 6, fontsize=14)
ax2.tick_params(axis='y', direction='in')
ax.set_title('(c) Subplot 3', y=-0.4, fontsize=14)

handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, fontsize=14, ncol=3, frameon=False, loc=[0, 0], bbox_to_anchor=[0.3, 1])

fig.tight_layout(pad=0, h_pad=1, w_pad=1)
# fig.savefig('fig/multi.pdf', transparent=True, bbox_inches='tight', pad_inches=0.1)
```
